```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
```

```python
crime=pd.read_csv('E:\\crime_data.csv')
crime.head()
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
```

```python
df_norm.head()
```

```python
dendrogram=sch.dendrogram(sch.linkage(df_norm,method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity = 'euclidean', linkage = 'complete')
hc
```

```python
y_hc=hc.fit_predict(df_norm)
y_hc
```

```python
crime['h_clusterid']=hc.labels_
```

# Kmeans

```python
from sklearn.cluster import KMeans
```

```python
crime1=pd.read_csv('E:\\crime_data.csv')
crime1.head()
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(crime.iloc[:,1:])
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
crime1['Clust']=x
```

```python
crime1.head()
```

```python
crime1.iloc[:,1:5].groupby(crime1.Clust).mean()
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime2=pd.read_csv('E:\\crime_data.csv')
```

```python
crime2.info()
```

```python
df=crime2.iloc[:,1:5]
```

```python
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
dbscan.labels_
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([crime2,cl],axis=1)
```
